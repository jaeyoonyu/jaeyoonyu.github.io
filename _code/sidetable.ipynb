```python
import sidetable
import pandas as pd
```

```python
url= 'https://github.com/chris1610/pbpython/blob/master/data/school_transform.csv?raw=True'
df = pd.read_csv(url, index_col=0)
df.head()
```

----------------

## Frequency

### Cumulative N and percent based on frequencies

```python
df.stb.freq(['Region'])
```

### Cumulative N and percent based on a specific values

```python
df.stb.freq(['Region'], value='Award_Amount')
```

```python
df.stb.freq(['Region'], value='Award_Amount', style= True)
```

----------

## Aggregate minor obs when there are too many categories

### Display all

```python
df.stb.freq(['State']) # df.stb.freq() to build a frequency table
```

### Display major only

```python
df.stb.freq(['State'], thresh=30, other_label='Rest of states') # detailed results up to 30% of cumulative percents
```

```python
df.stb.missing()
```
