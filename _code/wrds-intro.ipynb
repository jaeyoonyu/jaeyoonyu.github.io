```python
import wrds
```

```python
conn=wrds.Connection()
```

```python
conn.list_libraries()
```

```python
conn.list_tables(library="comp")
```

```python
company=conn.get_table(library='comp', table='company', obs=5)
print(company.shape)
company.head()
```

```python
company=conn.get_table(library='comp', table='company', 
                    columns = ['conm', 'gvkey', 'cik'], obs=5)
print(company.shape)
company.head()
```

```python
apple= conn.raw_sql("""select permno, date, prc, ret, shrout 
                    from crsp.msf
                    where permno= 14593 and date >= '01/01/2020'""",
                  date_cols=['date'])
apple
```
