# Usage of DataFrame.cut()
- Use cut when you need to segment and sort data values into bins. This function is also useful for going from a continuous variable to a categorical variable. For example, cut could convert ages to groups of age ranges. Supports binning into an equal number of bins, or a pre-specified array of bins. 
- Syntax: <code>pandas.cut(x, bins, right=True, labels=None, retbins=False, precision=3, include_lowest=False, duplicates='raise', ordered=True)</code>
- Description from [pandas documentation](https://pandas.pydata.org/docs/reference/api/pandas.cut.html)

------

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('compustat_from_1990.csv')
df.head()
```

```python
df_2020=df[df.fyear==2020].copy()
print(df_2020.shape)
df_2020.hist('mv')
plt.show()
```

```python
mv_max=max(df_2020.mv)
```

```python
bins=list(range(0,1000 + 1,100)) + [mv_max]
df_2020['mv_group']= pd.cut(df_2020['mv'], bins=bins)
df_2020.groupby('mv_group').count()['conm'].plot(kind='bar', figsize= (10,6))
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
bins=list(range(0,1000 + 1,100))
df_2020['mv_group']= pd.cut(df_2020['mv'], bins=bins)
df_2020.groupby('mv_group').count()['conm'].plot(kind='bar', figsize= (10,6))
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
bins=list(range(0,100 + 1,3))
df_2020['mv_group']= pd.cut(df_2020['mv'], bins=bins)
df_2020.groupby('mv_group').count()['conm'].plot(kind='bar', figsize= (10,6))
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
